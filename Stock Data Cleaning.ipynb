```python
import pandas as pd
```

```python
stocks = pd.read_csv('all_stocks_5yr.csv')
```

```python
stocks
```

```python
stocks['date'] = pd.to_datetime(stocks['date'])
```

```python
start_date = "2017-1-1"
end_date = "2017-12-31"

after_start_date = stocks["date"] >= start_date
before_end_date = stocks["date"] <= end_date
between_two_dates = after_start_date & before_end_date
stocks_2017 = stocks.loc[between_two_dates]
```

```python
stocks_2017
```

```python
stocks_2017['quarter'] = pd.PeriodIndex(stocks_2017.date, freq='Q')
```

```python
stocks_2017
```

```python
stocks_2017_grouped = stocks_2017.groupby(['Name', 'quarter']).agg({'close': ['mean', 'min', 'max', 'std']})
stocks_2017_grouped.columns = ['price_mean', 'price_min', 'price_max', 'price_std']
stocks_2017_grouped = stocks_2017_grouped.reset_index()
```

```python
stocks_2017_grouped
```

```python
main = stocks_2017_grouped.pivot_table(index='Name', columns='quarter', values=['price_mean', 'price_min', 'price_max', 'price_std'])
```
