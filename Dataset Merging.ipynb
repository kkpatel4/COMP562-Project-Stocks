```python
import pandas as pd
```

```python
prices = pd.read_csv('prices.csv')
sp = pd.read_csv('sp_500.csv')
finance = pd.read_csv('2017_Financial_Data.csv')
founded = pd.read_csv('founded.csv')
```

```python
comb = sp.merge(prices, on='Symbol')
```

```python
comb
```

```python
value_list = list(comb['Symbol'])
```

```python
included = finance['Symbol'].isin(value_list)
```

```python
fil_finance = finance[included]
```

```python
fil_finance
```

```python
main = comb.merge(fil_finance, on='Symbol')
```

```python
main
```

```python
final = main.merge(founded, on='Symbol')
```
